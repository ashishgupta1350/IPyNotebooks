### Q1. Find number of job openings in  _Bangalore,India_  and  in  _Seattle,US_

```python
job_location_India = 'Bangalore'
job_location_US = 'Seattle'
lines=0
openings_Bangalore=0
openings_Seattle=0
with open('amazon_jobs_dataset.csv') as csv_file:
    job_reader = csv.reader(csv_file)
 
    # Skipping the zeroth row which contains the column headers
    next(job_reader)    
    
    # Every row is a list and row[2] is the index for the feature 'Job Location'   
    for row in job_reader:
        
        # Job openings in Bangalore,India
        if job_location_India in row[2]:
            openings_Bangalore+=1
        # Job opening in Seattle,US
        if job_location_US in row[2]:
            openings_Seattle+=1
        lines+=1
```

```python
print("Number of job openings in Bangalore: ",openings_Bangalore)
print("Number of job openings in Seattle: ",openings_Seattle)
```

### Q2. What are the total number of job openings related to _Computer Vision_  ?

```python
title = 'Computer Vision'
lines=0
openings_cv=0
mydict=dict()
with open('amazon_jobs_dataset.csv') as csv_file:
    job_reader = csv.reader(csv_file)
 
    # Skipping the zeroth row which contains the column headers
    next(job_reader)    
    
    # Every row is a list and row[1] is the index for the feature 'Job Title'   
    for row in job_reader:
        # Job profile 
        if title in row[1]:
            mydict[row[1]]=mydict.get(row[1],0)+1 # checking if item is present in dictionary , if not then add it in 
            openings_cv+=1
        lines+=1
```

### Q3. Total number of job openings after Feb 2 , 2018 ?

#### Date Conversion in Python [ REALLY IMPORTANT CONCEPT FOR THIS QUESTION]
#### Link refered : https://www.guru99.com/date-time-and-datetime-classes-in-python.html
#### Summary of the link : https://www.quora.com/How-do-I-convert-different-date-formats-into-single-date-format-using-python

Date to convert :

**Case 1** : "2018-3-28"

**Case 2** : "Wed, 28, March, 18"

**Case 3** : "Wednesday, 28, March, 18"
 

Output should be : 03/28/2018

**Some notations :**
(%y/%Y â€“ Year), (%a/%A- weekday), (%b/%B- month), (%d - day of month)

**Example : %a is for Wed and %A is for Wednesday , %Y is for 2018 and %y is for 18** 

```python
# Import statement
import datetime
 
# First type of Date and it's format
inputDate0 = "2018-3-28"
DateFormat0 = "%Y-%m-%d"

# Second type of Date and it's format
inputDate1 = "Wed, 28, March, 18"
DateFormat1 = "%a, %d, %B, %y"
 
# Third type of Date and it's format
inputDate2 = "Wednesday, 26, March, 18"
DateFormat2 = "%A, %d, %B, %y"
 
outPutDateFormat = "%m/%d/%Y"
 
 # Converting input date to Date type object
date0 = datetime.datetime.strptime(inputDate0 , DateFormat0 )
date1 = datetime.datetime.strptime(inputDate1 , DateFormat1 )
date2 = datetime.datetime.strptime(inputDate2 , DateFormat2 )
 
print(date0)
 # Print different combination of dates as per output format
output1=datetime.date.strftime(date0, outPutDateFormat)
output2=datetime.date.strftime(date1, outPutDateFormat)
output3=datetime.date.strftime(date2, outPutDateFormat)

print('\n',output1,'\n',output2,'\n',output3,'\n')

print(output2>output3)
print(output2<output3)
```

```python
given_date = '2018/02/02' # 2nd February 2018 in (%Y/%m/%d format)
lines=0
count=0
mydict=dict()
outPutDateFormat = "%Y/%m/%d"

with open('amazon_jobs_dataset.csv') as csv_file:
    job_reader = csv.reader(csv_file)
 
    # Skipping the zeroth row which contains the column headers
    next(job_reader)    
    
    # Every row is a list and row[3] is the index for the feature 'Job posting date'   
    for row in job_reader:
        
        inputDate = row[3]
        # All jobs are in format : Month  Date, Year
        DateFormat = "%B %d, %Y"
        # Explained in the above article what does 'B'/'b'/'D'/'d'/'Y'/'y'/'M'/'m' all stand for..
     
        # Converting input date to Date type object
        datetime_object = datetime.datetime.strptime(inputDate,DateFormat)

        # Print different combination of dates as per output format
        converted_date = datetime.date.strftime(datetime_object, outPutDateFormat)
        
        print(inputDate," is converted to :",converted_date," and is compared with ",given_date)
        
        if converted_date > given_date:
            print('Yes, it is after 2nd Feb 2018 \n')
            count+=1
        else:
            print('No, it is NOT after 2nd Feb 2018 \n')
        lines+=1
```

```python
print('Job posting after 2nd February , 2018 : ',count)
```

```python
print(round(count/lines * 100 ,3) , '%  of job posting after 2nd Feb , 2018')
```
