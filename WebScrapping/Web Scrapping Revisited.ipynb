```python
import bs4
import urllib.request 
```

```python
sauce = urllib.request.urlopen('https://pythonprogramming.net/parsememcparseface/').read()
```

```python
type(sauce)
```

```python
soup = bs4.BeautifulSoup(sauce,'lxml') # lxml is the parser
```

```python
type(soup)
```

```python
soup
```

## For Extracting Content

```python
print(soup.title.text) # Only extracts for the first title tag 
```

```python
print(soup.p.text) # Only extracts for the first paragraph tag
```

```python
print(soup.find_all('p')) # Extracts all 'p' tags
```

```python
for para in soup.find_all('p'):
    print(para.get_text())
    print()
```

```python
print(soup.prettify())
```

```python
for div in soup.find_all('div',class_='navbar-fixed'): # For div tag under this class
    print(div.text)
    print()
```

## For Extracting Links :

```python
for link in soup.find_all('a'):
    print(link.get('href')) # get a particular attribute of a tag
```

## For Extracting Tables 

```python
table = soup.table # same as soup.find('table')
```

```python
table
```

```python
table_rows = table.find_all('tr')
```

```python
for tr in table_rows:
    td = tr.find_all('td') # 'td' is the feature value for a particular data point i.e each row's feature value
    row =[row.text for row in td] # We are going to extract the text from all the 'td' that we found in a row
    print(row)
```

```python
wiki_waca = urllib.request.urlopen('https://en.wikipedia.org/wiki/List_of_international_cricket_centuries_at_the_WACA_Ground').read()
```

```python
wiki_waca_soup = bs4.BeautifulSoup(wiki_waca,'lxml')
```

```python
table = wiki_waca_soup.find('table')
```

```python
table_rows = table.find_all('tr')
```

```python
rows = []
for tr in table_rows:
    tds = tr.find_all('td')
    row = []
    for td in tds:
        row.append(td.text.strip())
    rows.append(row)
```

```python
heads = []
# For extracting the header from the table
ths = table_rows[0].find_all('th') # table headers
for th in ths:
    heads.append(th.text.strip())
```

```python
rows[0] = heads # Making this as the table header column
```

```python
rows[0]
```

```python
type(rows[0])
```

```python
df = pd.DataFrame(rows,columns=heads)
```

```python
df.drop(0,inplace=True)
```

```python
df
```

```python
df.to_csv('waca_test_centuries', sep=',',index=False) 
# When you are storing a DataFrame object into a csv file using the to_csv method
# You probably wont be needing to store the preceding indices of each row of the DataFrame object.
# You can avoid that by passing a False boolean value to index parameter.
```
