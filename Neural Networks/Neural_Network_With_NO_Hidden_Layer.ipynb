```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.array([[0,0,0,1]]).T
X.shape,Y.shape # 2 features for each data ppint 
```

```python
def sig(z):
    return 1 / (1 + np.exp(-z))
```

```python
def derivativeSig(z):
    return sig(z)*(1-sig(z))
```

### We have no hidden layer just 1 output layer 

```python
# no hidden layer weights
weights = 2 * np.random.random((2,1)) - 1
print(weights)
bias = 2 * np.random.random(1) - 1
print(bias)
lr = 0.1
```

```python
# forward propogation without hidden layer
for iter in range(10000):
    output0 = X 
    output = sig(np.dot(output0,weights)+bias)
    first_term = output - Y
    input_for_last_layer = np.dot(output0,weights) + bias
    second_term = derivativeSig(input_for_last_layer)
    first_two = first_term * second_term
    #print(first_term.shape,second_term.shape,first_two.shape)
    changes = np.dot(output0.T,first_two)
    weights = weights - lr*changes
    bias_changes = np.sum(first_two)
    bias = bias - lr*bias_changes
    
output = sig(np.dot(output0,weights)+bias)
weights,bias,output
```
